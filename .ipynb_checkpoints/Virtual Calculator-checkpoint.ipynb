```python
import cv2 
from cvzone.HandTrackingModule import HandDetector
import time 
```

```python
class Button:
    def __init__(self,pos,width,height, value):
        self.pos = pos 
        self.width = width 
        self.height = height 
        self.value = value 
        
        
    def Draw(self,img):
        cv2.rectangle(img, self.pos,(self.pos[0]+self.width,self.pos[1]+self.height),(225,225,225),cv2.FILLED)
        cv2.rectangle(img, self.pos,(self.pos[0]+self.width,self.pos[1]+self.height),(50,50,50),3)
        cv2.putText(img,self.value,(self.pos[0] + 38, self.pos[1] + 60),cv2.FONT_HERSHEY_PLAIN,2,(50,50,50),2)    
    
    def checkClick(self,x,y):
        if self.pos[0] < x < self.pos[0] + self.width and self.pos[1] < y < self.pos[1] + self.height:
            cv2.rectangle(img, self.pos,(self.pos[0]+self.width,self.pos[1]+self.height),(255,255,255),cv2.FILLED)
            cv2.rectangle(img, self.pos,(self.pos[0]+self.width,self.pos[1]+self.height),(50,50,50),3)
            cv2.putText(img, self.value,(self.pos[0] + 20, self.pos[1] + 80),cv2.FONT_HERSHEY_PLAIN,5,(0,0,0),5)
            return True 
        else:
                return False 
    
# webcom 
cap =cv2.VideoCapture(0)
cap.set(3,1080) # width
# cap.set(4,400) # height
detector = HandDetector(detectionCon=0.8, maxHands=1)

#  create buttons 
buttonListValues = [
                       ['7' , '8' , '9' , '*'],
                       ['4' , '5' , '6' , '-'],
                       ['1' , '2' , '3' , '+'],
                       ['0' , '/' , '.' , '=']                       
                   ]
buttonList = []
for x in range(4):
    for y in range(4):        
        xpos = x*100 + 800 
        ypos = y * 100 +150 
        buttonList.append(Button((xpos,ypos),100,100,buttonListValues[y][x]))

# variables

myEquation = ''
delayCounter = 0 



while True:
    #     get image from webcom 
    success, image=cap.read()
    image=cv2.flip(image,1)
     
    #     detection of hand 
    hands, image = detector.findHands(image, flipType=False)
    
    #  Draw All Buttons 
    
    cv2.rectangle(image, (800,70),(800 + 400, 70 + 100),(225,225,225),cv2.FILLED)
    cv2.rectangle(image, (800,70),(800 + 400, 70 + 100) ,(50,50,50),3)
    for button in buttonList:        
        button.Draw(image)
        
    # Processing // check the hand 
    if hands:
        lmList = hands[0]['lmList']
        length,_, image = detector.findDistance(lmList[8], lmList[12], image)
        x,y = lmList[8]
        if length<50:
            for i, button in enumerate(buttonList):
                if button.checkClick(x,y) and delayCounter == 0:
                    myValue = buttonListValues[int(i%4)][int(i/4)]
                    if myValue == '=':
                        myEquation = str(eval(myEquation))
                    else:
                        myEquation += myValue                 
                    delayCounter = 1 
                
    # Avoid Duplicate 
    if delayCounter !=0:
        delayCounter +=1
        if delayCounter >10:
            delayCounter = 0 
    # Display the Result     
    cv2.putText(image,myEquation,(830, 130),cv2.FONT_HERSHEY_PLAIN,3,(50,50,50),3)
    
    #  display image 
    cv2.imshow("image",image)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        myEquation='' 
    
cv2.destroyAllWindows()
```
